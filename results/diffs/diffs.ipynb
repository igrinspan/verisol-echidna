```python
import json
import pandas as pd

f = open('diff_results_0_8_0.json')
diffs = json.load(f)
df_diffs = pd.DataFrame.from_dict(diffs)
```

```python
f = open('../times/verisol_times.json')
verisol_times = json.load(f)
df_verisol = pd.DataFrame.from_dict(verisol_times)
```

```python
df_diffs = pd.merge(df_verisol, df_diffs, on='Contract', how='outer')
```

```python
df_diffs['Diff Epa 500'] = None
df_diffs['Diff States 500'] = None
df_diffs['Diff Epa 50000'] = None
df_diffs['Diff States 50000'] = None
df_diffs['Diff Epa 500000'] = None
df_diffs['Diff States 500000'] = None
```

```python
# turn node difference and edge difference into integers
df_diffs['Node Difference'] = df_diffs['Node Difference'].astype('Int64', errors='ignore')
df_diffs['Edge Difference'] = df_diffs['Edge Difference'].astype('Int64', errors='ignore')
df_diffs.head()
```

```python
df_nueva = df_diffs
for index, row in df_diffs.iterrows():
    contract = row['Contract']
    test_limit = row['Test Limit']
    mode = row['Mode']
    nodes = row['Node Difference']
    edges = row['Edge Difference']
    
    if test_limit == 500:
        if mode == 'epa':
            df_nueva.at[index, 'Diff Epa 500'] = (nodes, edges)
        elif mode == 'states':
            df_nueva.at[index, 'Diff States 500'] = (nodes, edges)
    elif test_limit == 50000:
        if mode == 'epa':
            df_nueva.at[index, 'Diff Epa 50000'] = (nodes, edges)
        elif mode == 'states':
            df_nueva.at[index, 'Diff States 50000'] = (nodes, edges)
    elif test_limit == 500000:
        if mode == 'epa':
            df_nueva.at[index, 'Diff Epa 500000'] = (nodes, edges)
        elif mode == 'states':
            df_nueva.at[index, 'Diff States 500000'] = (nodes, edges)
```

```python
df_nueva.head()
```

```python
df_nueva.drop(columns=['Verisol Epa Time', 'Verisol States Time', 'Mode', 'Test Limit', 'Node Difference', 'Edge Difference'], inplace=True)
```
