```python
import json
import pandas as pd

f = open('echidna_times.json')
echidna_times = json.load(f)
df_echidna = pd.DataFrame.from_dict(echidna_times)
```

```python
f = open('verisol_times.json')
verisol_times = json.load(f)
df_verisol = pd.DataFrame.from_dict(verisol_times)
```

```python
df_joined = pd.merge(df_verisol, df_echidna, on='Contract', how='outer')
```

```python
df_joined['Echidna Epa Time 100'] = None
df_joined['Echidna States Time 100'] = None
df_joined['Echidna Epa Time 50000'] = None
df_joined['Echidna States Time 50000'] = None
df_joined['Echidna Epa Time 500000'] = None
df_joined['Echidna States Time 500000'] = None

df_nueva = df_joined
for index, row in df_joined.iterrows():
    contract = row['Contract']
    test_limit = row['test_limit']
    mode = row['mode']
    time_in_s = row['time_in_s']
    
    if test_limit == 100.0:
        if mode == 'e':
            df_nueva.at[index, 'Echidna Epa Time 100'] = time_in_s
        elif mode == 's':
            df_nueva.at[index, 'Echidna States Time 100'] = time_in_s
    elif test_limit == 50000.0:
        if mode == 'e':
            df_nueva.at[index, 'Echidna Epa Time 50000'] = time_in_s
        elif mode == 's':
            df_nueva.at[index, 'Echidna States Time 50000'] = time_in_s
    elif test_limit == 500000.0:
        if mode == 'e':
            df_nueva.at[index, 'Echidna Epa Time 500000'] = time_in_s
        elif mode == 's':
            df_nueva.at[index, 'Echidna States Time 500000'] = time_in_s
```

```python
df_nueva.drop(columns=['mode', 'test_limit', 'time_in_s'], inplace=True)
```

```python
a = df_nueva.groupby(['Contract'])
```

```python
resultado = a.first()
```
